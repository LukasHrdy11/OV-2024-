# Výpočet numerického řešení Schrödingerovi rovnice

```python
import math
import numpy as np
import pandas as pd
#redukovaná hmotnost v MeV
my = 938*938/(938+938)
E = np.arange(2,3,0.01)
#E = 2.22452
hbar = 1.054571628e-34
h = 0.01
r = np.arange(0,3,h)
l = 0
```

## Energie

```python
dfE = pd.DataFrame(E)
dfE = dfE.rename(columns={0:"E"})
dfE
```

## r a V

```python
dfrV = pd.DataFrame(r)
dfrV["V"] = 200*pow(math.e, -1.487*pow(dfrV[0],2)) - 178*pow(math.e, -0.639*pow(dfrV[0], 2))
dfrV = dfrV.rename(columns={0:"r"})
dfrV
```

```python
import matplotlib.pyplot as plt
dfrV.plot(x = 'r', y = 'V')

plt.show()
```

## k

```python
dfk = pd.DataFrame()
V_radky = dfrV.shape[0]
E_radky = dfE.shape[0]
k_hodnoty = []
for i in range(0,E_radky):
    for z in range (0,V_radky):
        E = dfE.iloc[i,0]
        V = dfrV.iloc[z,1]
        k = 2*my/pow(hbar, 2)*(E - V)
        k_hodnoty.append(k)
    dfk[f"E{i}"] = k_hodnoty
    k_hodnoty = []
```

### V každé buňce jsou vypočteny hodnoty k pro Energii a radiální vzdálenost

```python
dfk
```

## Výpočet u

```python
dfu = pd.DataFrame()
u0 = 0
u1 = pow(h, (l+1))
for i in range(0,E_radky):
    u0 = 0
    u1 = pow(h, (l+1))
    u_hodnoty = [u0,u1]
    for z in range (0,298):    
        k0 = dfk.iloc[z,i]
        k1 = dfk.iloc[z+1,i]
        k2 = dfk.iloc[z+2,i]
        u2 = (2*(1-5/12*pow(h,2)*k1)*u1-(1+1/12*pow(h,2)*k0)*u0)/(1+pow(h,2)/12*k2)
        u_hodnoty.append(u2)
        u0 = u1
        u1 = u2
    dfu[f"E{i}"] = u_hodnoty
    
```

#### V každé buňce je vypočtena radiální vlnová funkce pro Energii a radiální vzdálenost. Sloupce jsou energie a řádky jsou vzdálenost.
